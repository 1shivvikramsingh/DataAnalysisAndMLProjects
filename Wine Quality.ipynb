```python
plt.figure(figsize=(8,6))
ax = sns.countplot(x='quality', data=wine, color='orange')
ax.set(title='Histogram For Wine Quality', xlabel='', ylabel='', yticklabels=[])
ax.tick_params(left=False)
for i in ax.patches:
    ax.annotate(i.get_height(),
                xy=(i.get_x() + i.get_width()/2, i.get_height()),
                xytext=(0,10),
                textcoords='offset points',
                ha='center',
                size=10)
plt.tight_layout()
```
